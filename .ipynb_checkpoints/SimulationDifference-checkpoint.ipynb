# Estimating Detector Resolution
This notebook describes how we determine the detector resolution by calculating the difference between the simulated true momenta and just simulated momenta. It is expected that these results should be consistent with those in the LHCb detector paper given at the start of the project.

## Getting the Data
The first step to achieving the above is to import the data from the ROOT file and extract the variables we need into a Pandas dataframe. Using Pandas over numpy is a somewhat arbitrary choice though mostly due to ease-of-use and familiarity. Note here we are doing the momentum not the position resolution.

```python
import uproot as up
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fName="/disk/moose/lhcb/djdt/Lb2L1520mueTuples/MC/2016MD/100FilesCheck/job185-CombDVntuple-15314000-MC2016MD_100F-pKmue-MC.root"
```

```python
with up.open(fName + ":DTT1520me/DecayTree") as f:
    df = f.arrays(["Lb_PX", "Lb_TRUEP_X", "Lb_PY", "Lb_TRUEP_Y", "Lb_TRUEP_Z", "Lb_PZ"], library="pd")
    df.columns = ['Px', 'TPx', 'Py', 'TPy', 'Pz', 'TPz']
print(df.head())
```

Great, so now we have the initial data that we need - neatly organised in a Pandas dataframe. The columns were renamed to make indexing them later a bit easier. We can now calculate the difference in each co-ordinate to get a momentum resolution across each axis and then one for the total difference accounting for all directions. 

```python
data = df.copy()
data['dPx'] = data['Px'] - data['TPx']
data['dPy'] = data['Py'] - data['TPy']
data['dPz'] = data['Pz'] - data['TPz']
data.drop(['Px', 'TPx', 'Py', 'TPy', 'Pz', 'TPz'], axis=1, inplace=True)
print(data.head())
```

Now we can plot each of these changes $\delta i=i_{measured}-i_{true}$ where $i$ is either $x$, $y$ or $z$. We plot these changes as a histogram.

```python
fig, ax = plt.subplots(1, 3, figsize=(16, 4))
bins = 400
x = ax[0].hist(data.loc[(data['dPx'] >= -4000) & (data['dPx'] <= 4000)]['dPx'], bins=bins)
ax[0].set_xlabel(r'$\delta x$ $(mm)$')
y = ax[1].hist(data.loc[(data['dPy'] >= -4000) & (data['dPy'] <= 4000)]['dPy'], bins=bins)
ax[1].set_xlabel(r'$\delta y$ $(mm)$')
z = ax[2].hist(data.loc[(data['dPz'] >= -8000) & (data['dPz'] <= 8000)]['dPz'], bins=bins)
ax[2].set_xlabel(r'$\delta z$ $(mm)$')
fig.suptitle('Difference between measured and true value for momentum')
fig.supylabel('Count')
plt.show()
# Some regions of the data have been cut off as the bins of ~1 result very far out caused the plots to be difficult to read,
# these data are not discarded when we pass it through the fitting routines. 
```

Visualising the z-momentuum of the $\lambda_B$ a little more let us plot a histogram of the true versus the 'measured' momentum to see if there is anything interesting at play.

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
bins = 500
ax.hist(df['TPz'], bins=bins, label='Truth Data', alpha=0.7)
ax.hist(df['Pz'], bins=bins, label='Measured Data', alpha=0.7)
ax.set_xlabel(r'Measured Momentum of $\Lambda_B$ (eV)')
ax.set_xscale('log')
ax.set_xlim(1000, 1e6)
plt.legend()
plt.show()
```

```python
print(sorted(df['Pz'].to_list()))
```

This is indeed very strange behaviour! Why is there such a large frequency of negative momentum measurements. Obviously this just means the particles are travelling in the opposite direction however this is not seen in the truth level data.

```python
df['P'] = np.arccos(df['Pz']/np.sqrt(df['Px']**2 + df['Py']**2 + df['Pz']**2))
df['TP'] = np.arccos(df['TPz']/np.sqrt(df['TPx']**2 + df['TPy']**2 + df['TPz']**2))

fig, ax = plt.subplots(1, 1, figsize=(12, 7))
bins = 500
ax.hist(df['P'], bins=bins, label='Truth Momentum', alpha=0.7)
ax.hist(df['TP'], bins=bins, label='Measured Data', alpha=0.7)
ax.set_xlabel(r'$\cos\theta$')
fig.suptitle('Title')
ax.set_yscale('log')
fig.supylabel('Count')
plt.legend()
plt.show()
```

```python
large_rad = df.loc[df['P'] > 0.5]
print(large_rad.describe())
```

The detector appears to over-estimate the momentum more often than it under-estimates it. This behaviour follows that which was observed in the z-direction. Why does the detector seemingly detect more often than not, a higher z-momentum than the true value?

```python
xbins = []
bin_width = x[1][1] - x[1][0]
for i in range(bins):
    bin_middle = x[1][i]+(bin_width/2)
    xbins.append(bin_middle)

ybins = []
bin_width = y[1][1] - y[1][0]
for i in range(bins):
    bin_middle = y[1][i]+(bin_width/2)
    ybins.append(bin_middle)

zbins = []
bin_width = z[1][1] - z[1][0]
for i in range(bins):
    bin_middle = z[1][i]+(bin_width/2)
    zbins.append(bin_middle)
```

## Analysis of Histogram Results
As can be seen above the errors in the $x$ and $y$ co-ordinate directions seem roughly equivalent and seem like they are roughly normally distributed. However in the z-direction the errors are much more spread out and seem to have a positive skew on loose inspection. Considering that the beam axis is the z-axis it is unsuprising that there is no difference between $x$ and $y$ since these should be symmetric. We can try to further analyse these results by quantifying them with some fits, we will try to fit a variety of functions using a basic least-squares fitting routine.

```python
from scipy.special import erf, wofz
from scipy.stats import crystalball
import scipy.optimize as opt

def voigt(x, alpha, gamma, shift):
    sigma = alpha/np.sqrt(2*np.log(2))
    return np.real(wofz(((x-shift) + 1j*gamma)/sigma/np.sqrt(2)))/sigma/np.sqrt(2*np.pi)

def normal(x, mu, s):
    # Where mu is the mean and s is the standard deviation
    return (1/np.sqrt(2 * np.pi * s**2)) * np.exp(-(x-mu)**2/s**2)
```

Now let us try and fit the $x$ and $y$ graphs using some of the functions given above and see which works best. We will use the scipy optimize package to do this quickly for now.

```python
p0 = [0, 400] # Initial guesses for the parameters
popt, pcov = opt.curve_fit(normal, np.array(xbins), x[0], p0)
# perr = np.sqrt(np.diag(pcov))
# Use sigma to provide an error in your data!
```

### Voigt Profile Fit

```python
voigt_p0 = [1, 0.5, 0]
vpopt, vcov = opt.curve_fit(voigt, np.array(xbins), x[0], voigt_p0)
# vperr = np.sqrt(np.diag(pcov))
```

### Crystal Ball Fit

```python
p2 = [2, 3, 0, 2]
popt2, pcov2 = opt.curve_fit(crystalball.pdf, np.array(xbins), x[0], p2)
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(xbins, normal(xbins, popt[0], popt[1]), label=rf'$\mathcal{{N}}\sim({popt[0]:.3f}, {popt[1]:.3f})$', alpha=0.7)
ax.plot(xbins, voigt(xbins, vpopt[0], vpopt[1], vpopt[2]), label=rf'$\mathcal{{V}}\sim({vpopt[0]:.3f}, {vpopt[1]:.3f}, {vpopt[2]:.3f})$', alpha=0.7)
ax.plot(xbins, crystalball.pdf(xbins, *popt2), label=rf'$\mathcal{{C}}\sim({popt2[0]:.3f}, {popt2[1]:.3f}, {popt2[2]:.3f}, {popt2[3]:.3f})$', alpha=0.7)
ax.plot(xbins, x[0], label='Original Data')
ax.set_xlabel(rf'Difference, $\Delta p_x=p_{{x, measured}}-p_{{x, true}}$')
ax.set_ylabel('Normalised Frequency')
#ax.set_xlim((100, 200))
ax.set_title('Fits onto the Difference Plot')
ax.legend(prop={'size': 14})
#plt.savefig('plots/BackgroundPolyFit.png')
plt.show()
```

This is interesting - nothing seems to be responding very well to fitting the x (and equivalently y) momentum measurement errors. The Gaussian is by far the worser fit. Oddly when normalising the frequency of the histogram the fits seem to do better.

```python
p0 = [2, 3, 0, 2] # Initial guesses for the parameters
popt, pcov = opt.curve_fit(crystalball.pdf, np.array(zbins), z[0], p0)
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(zbins, crystalball.pdf(zbins, *popt), label=rf'$\mathcal{{C}}\sim({popt[0]:.3f}, {popt[1]:.3f})$', alpha=0.7)
ax.plot(zbins, z[0], label='Original Data')
ax.set_xlabel('Difference')
ax.set_ylabel('Normalised Frequency')
#ax.set_xlim((100, 200))
ax.set_title('Fits onto the Difference Plot')
ax.legend(prop={'size': 14})
#plt.savefig('plots/BackgroundPolyFit.png')
plt.show()
```
