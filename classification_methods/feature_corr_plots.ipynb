# Feature Correlation
We now seek to analyse a fair number of features for any sorts of correlation.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import mplhep as hep
#plt.style.use([hep.style.ROOT, hep.style.firamath])
```

## Correlation Matrix
We need to first load the data and compute its correlation matrix.

```python
version = '7.0.2'
df = pd.read_csv(f'../data_files/{version}/all.csv', index_col=[0])
df.drop(['Lb_M', 'IsSimulated'], axis=1, inplace=True)
```

```python
corr = df.corr()
# Compute the correlation matrix

mask = np.triu(np.ones_like(corr, dtype=bool))
# Create a mask for the upper triangle of the array
```

## Plotting Correlation
Now we have all we need to plot the figure.

```python
fig, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
# Generate a figure and diverging colour map

sns.heatmap(corr, cmap=cmap, vmax=.6, center=0, square=True, linewidth=.5, cbar_kws={"shrink": .5}, mask=mask)
ax.set_xlabel('Features')
ax.set_ylabel('Features')
plt.savefig(f'../images/7.0.2_correlation.png', dpi=800)
plt.tight_layout()
plt.show()
```
