# Feature Correlation
We now seek to analyse a fair number of features for any sorts of correlation.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sys
sys.path.append('../')
from dataflow import Flow
```

## Correlation Matrix
We need to first load the data and compute its correlation matrix.

```python
version = '5.0.2'
fname = f'../neural_network/csv/preselected_feat_v{version}_equalRatio_restrictedSideband.csv'
data = Flow(None, None, None, fname)
df = data.get_combined_data()
```

```python
corr = df.corr()
# Compute the correlation matrix

mask = np.triu(np.ones_like(corr, dtype=bool))
# Create a mask for the upper triangle of the array
```

## Plotting Correlation
Now we have all we need to plot the figure.

```python
fig, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
# Generate a figure and diverging colour map

sns.heatmap(corr, cmap=cmap, vmax=.6, center=0, square=True, linewidth=.5, cbar_kws={"shrink": .5})
ax.set_xlabel('Features')
ax.set_ylabel('Features')
plt.savefig(f'../../images/feature_correlation_plot_{version}.png')
plt.show()
```
